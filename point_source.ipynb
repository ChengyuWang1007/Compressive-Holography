```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lambda_=0.633# (um)
dim_X = 512
dim_Y = 512
```

```python
## coordinates of the point source
x = 250
y = 248

## pixel size
pixel_size = 5.2 #(um)

##depth
z = 24000#(um)
```

```python
## Intensity of the point source
I = 1
```

```python
xv, yv = np.meshgrid(np.arange(512), np.arange(512))
```

```python
r = np.sqrt(((x - xv) * pixel_size)**2 + ((y - yv) * pixel_size)**2 + z**2)
```

```python
Y = I * np.exp(1j * 2 * np.pi / lambda_ * r)/r
```

```python
plt.imshow(np.abs(Y))
```
